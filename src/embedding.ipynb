```python
import os
from langchain_core.documents import Document
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
DATA_PATH = "/content/drive/MyDrive/Colab Notebooks/Capstone/Data"
file_list = os.listdir(DATA_PATH)
```

```python
file_list[:4]
```

```python
loaded_docs = []


for i in range(4):
    loader = PyPDFLoader(DATA_PATH +"/" + file_list[i])
    doc = loader.load()
    loaded_docs.append(doc)

```

```python
for i in range(4):
    print("Loaded PDF Content\n")
    print(f"{loaded_docs[i][0].page_content[:200]}\n")
    print(f"{loaded_docs[i][0].metadata}\n")
```

```python
for page in loaded_docs[2]:
    print(f"Length - {len(page.page_content)}")
    print(f"{page.metadata}\n")
```

```python
all_splitts_docs = []

for i in range(4):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000, chunk_overlap=200, add_start_index=True
    )
    all_splits = text_splitter.split_documents(loaded_docs[i])
    all_splitts_docs.append(all_splits)
    print(len(all_splits))
```

```python
len(all_splitts_docs)
```

```python
for i in range(4):
    print(f"Document {i} splitted into {len(all_splitts_docs[i])}\n")
```

```python
all_splitts_docs[1]
```

```python
for split in all_splitts_docs[2]:
    print(f"Length - {len(split.page_content)}")
    print(f"{split.metadata}\n")
```

```python
print(all_splits[1])
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
vector_1 = embeddings.embed_query(all_splits[0].page_content)
vector_2 = embeddings.embed_query(all_splits[1].page_content)

assert len(vector_1) == len(vector_2)
print(f"Generated vectors of length {len(vector_1)}\n")
print(vector_1[:10])
```

```python
splits_docs = []
for doc in all_splitts_docs:
    for split in doc:
        splits_docs.append(split)
```

```python
print(len(splits_docs))
```

```python
from langchain_core.vectorstores import InMemoryVectorStore


vector_store = InMemoryVectorStore(embeddings)
```

```python
ids = vector_store.add_documents(documents=splits_docs)

```

```python
print(len(ids))
```

```python
results = vector_store.similarity_search_with_score(
    "How faculty can purchase parking permit?"
)
print(len(results))

for i in range(4):
    print(f"Result {i+1} with Score {results[i][1]} \n\n {results[i][0]}\n\n")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://kb.uconn.edu/space/PAR/10894836117/Parking+Portal"
atlas_url = "https://uconn.atlassian.net/wiki/spaces/PAR/pages/10894836117/Parking+Portal"


response = requests.get(atlas_url)

if response.status_code == 200:
    print(f"Successfully fetched the page: {url}")
else:
    print(f"Failed to retrieve the page: {url}, Status code: {response.status_code}")
    exit()

with open("Parking_alt.html",'wb') as f:
    f.write(response.content)



```

```python
# Parse the page content with BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Assuming that articles or knowledge base entries are stored in specific HTML elements
# Replace the selector below with the actual structure from UConn's knowledge base
articles = soup.find_all('div', class_='article')  # Adjust based on the HTML structure of the KB

# Create a list to hold the scraped data
scraped_data = []

for article in articles:
    title = article.find('h2').text.strip() if article.find('h2') else 'No Title'
    link = article.find('a')['href'] if article.find('a') else 'No Link'
    summary = article.find('p').text.strip() if article.find('p') else 'No Summary'

    # Store the data in a dictionary
    scraped_data.append({
        'Title': title,
        'Link': link,
        'Summary': summary
    })

# Convert the data into a pandas DataFrame for easy analysis
df = pd.DataFrame(scraped_data)

# Show the first few rows of the DataFrame
print(df.head())

# Optionally, save the data to a CSV file
df.to_csv("uconn_kb_scraped_data.csv", index=False)

print("Data saved to uconn_kb_scraped_data.csv")
```
